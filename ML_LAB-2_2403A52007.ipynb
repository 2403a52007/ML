TASK-1

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/Heart_Disease_Prediction.csv')

print('First 5 rows of the dataset:')
display(df.head())

print('\nLast 5 rows of the dataset:')
display(df.tail())

print(f'\nNumber of rows and columns: {df.shape}')

print('\nColumn names:')
display(df.columns)
```

TASK-2

```python
print('Dataset Information:')
df.info()

numerical_features = df.select_dtypes(include=np.number).columns.tolist()
categorical_features = df.select_dtypes(exclude=np.number).columns.tolist()

print(f'\nNumerical Features: {numerical_features}')
print(f'Categorical Features: {categorical_features}')

print('\nData Types of all columns:')
display(df.dtypes)

print('\nMissing values per column:')
display(df.isnull().sum())
```

TASK-3

```python
print('Statistical Summary of the dataset:')
display(df.describe())

print('\nMean Age:', df['Age'].mean())
print('Median Age:', df['Age'].median())
print('Standard Deviation of Age:', df['Age'].std())

print('\nMinimum Cholesterol:', df['Cholesterol'].min())
print('Maximum Cholesterol:', df['Cholesterol'].max())
```

TASK-4

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Count number of patients with and without heart disease
heart_disease_counts = df['Heart Disease'].value_counts()
print('Number of patients with and without Heart Disease:')
display(heart_disease_counts)

# Task 2: Plot histogram for age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=10, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Task 3: Plot histogram for cholesterol levels
plt.figure(figsize=(10, 6))
sns.histplot(df['Cholesterol'], kde=True, bins=15, color='lightcoral')
plt.title('Distribution of Cholesterol Levels')
plt.xlabel('Cholesterol')
plt.ylabel('Number of Patients')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

TASK-5

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df[numerical_features].corr()

print('Correlation Matrix:')
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

high_corr = correlation_matrix[((correlation_matrix.abs() > 0.7) & (correlation_matrix.abs() < 1.0))].stack().sort_values(ascending=False)
print('\nHighly Correlated Variable Pairs (Absolute correlation > 0.7):')
display(high_corr.to_frame(name='Correlation Value'))

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Max HR', data=df, hue='Heart Disease', palette='viridis', alpha=0.7)
plt.title('Scatter Plot of Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.grid(True, alpha=0.5)
plt.show()
```
