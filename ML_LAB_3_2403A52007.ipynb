Plot regression line with data points.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/content/Salary_dataset.csv')
df_cleaned = df.drop('Unnamed: 0', axis=1)
X = df_cleaned[['YearsExperience']]
y = df_cleaned['Salary']
X_reshaped = X.values.reshape(-1, 1)
y_reshaped = y.values.reshape(-1, 1)
slope = 0.0
intercept = 0.0
learning_rate = 0.01
epochs = 1000
n_samples = len(y_reshaped)
print('Starting Model Training with Gradient Descent...')
for i in range(epochs):
    y_pred_current = slope * X_reshaped + intercept

    d_slope = (2/n_samples) * np.sum(X_reshaped * (y_pred_current - y_reshaped))
    d_intercept = (2/n_samples) * np.sum(y_pred_current - y_reshaped)
    slope = slope - learning_rate * d_slope
    intercept = intercept - learning_rate * d_intercept

    if i % 200 == 0:
        cost = (1/n_samples) * np.sum(np.square(y_pred_current - y_reshaped))
        print(f'Epoch {i}, Cost: {cost:.2f}, Slope: {slope:.2f}, Intercept: {intercept:.2f}')

print('\n--- Training Complete ---')
print(f'Final Trained Slope: {slope:.2f}')
print(f'Final Trained Intercept: {intercept:.2f}')
final_y_pred = slope * X_reshaped + intercept

plt.figure(figsize=(10, 6))
plt.scatter(X_reshaped, y_reshaped, color='blue', label='Actual Data')
plt.plot(X_reshaped, final_y_pred, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs. Salary with Regression Line (Trained Model)')
plt.legend()
plt.grid(True)
plt.show()
```

 Interpret the results

```python
import matplotlib.pyplot as plt

# Use predictions from the trained model
y_pred = final_y_pred # Assuming final_y_pred is available from the trained model
residuals = y_reshaped - y_pred

plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, color='purple', alpha=0.7)
plt.hlines(y=0, xmin=y_pred.min(), xmax=y_pred.max(), colors='red', linestyles='--')
plt.xlabel('Predicted Salary')
plt.ylabel('Residuals (Actual - Predicted)')
plt.title('Residual Plot of Trained Model') # Updated title
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('Residual Value')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals for Trained Model') # Updated title
plt.grid(True)
plt.show()
```
